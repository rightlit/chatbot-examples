```python
import pandas as pd

df = pd.read_excel('./wellness_dialog_dataset.xlsx')
df
```

```python
import pandas as pd
import numpy as np

Q_dic = {}
A_dic = {}
for i, catg in enumerate(df['구분']):
    if(catg in Q_dic):
        Q_dic[catg].append(df['유저'][i])
        if(pd.isna(df['챗봇'][i]) == False):
            A_dic[catg].append(df['챗봇'][i])
    else:
         Q_dic[catg] = []
         A_dic[catg] = []
```

```python
#print(Q_dic)
```

```python
#print(A_dic)
```

```python
import pandas as pd
import re
import csv

cnt = 0
#df = pd.read_excel('./wellness_dialog_dataset.xlsx')
QA_data = list()
for catg in Q_dic.keys():
    q_data = Q_dic[catg]
    a_data = A_dic[catg]
    for i, q_text in enumerate(q_data):
        for j, a_text in enumerate(a_data):
            qa_text = q_text + '    ' + a_text
            #qa_text = qa_text.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
            #qa_text = qa_text.replace("[ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
            #qa_text = re.sub("[ㅠㅜ]","", qa_text)
            QA_data.append(qa_text)
            cnt = cnt + 1
            if(cnt < 1000):
                print(catg, i,j,qa_text)

#df_qa = pd.DataFrame(data=QA_data, columns=['QA'])
#df_qa['QA'] = QA_data
#df_qa['QA'].to_csv('./chatbot_dialog_wellness.txt', index=False, header=False)
#df_qa['QA'].to_csv('./chatbot_dialog_wellness.txt', index=False, header=False, sep='    ')

with open("chatbot_dialog_wellness.txt", "w") as f:
    for line in QA_data:
        f.write(line + '\n')
```
